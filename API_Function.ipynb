```python
import requests
import pandas as pd
```

```python
def api_data_frame(lat,long):
    api_url4 = "https://api.openweathermap.org/data/2.5/forecast?lat={}&lon={}&appid=22ba7773328731b108fe5951ed3df034".format(lat,long)
    response1 = requests.get(api_url4)
    mm = response1.json()
    
    ##forecast data frame with three hours lag
    lst = []
    for i in mm['list']:
        frm1 = pd.DataFrame((i['main']).items(), columns=['Column', 'Value'])
        frm1['dates' ]= i['dt_txt']
        frm2 = frm1.pivot(index= 'dates', columns = "Column",values ="Value").reset_index()
        lst.append(frm2)


    fnl_frm = pd.concat(lst).rename_axis(None,axis=1).reset_index().drop('index',1)
    fnl_frm['dates'] = pd.to_datetime(fnl_frm['dates'])
    fnl_frm = fnl_frm.sort_values(['dates'])
    fnl_frm['Latitude'] = lat
    fnl_frm['Longitude'] = long
    return fnl_frm
```
